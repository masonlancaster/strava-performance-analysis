```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('strava-activities.csv')
```

```python
rides = df.loc[df['sport_type'] == 'Ride']
```

```python
runs = df.loc[df['sport_type'] == 'Run']
```

```python
# Longest Run in distance
runs.loc[runs['distance'].idxmax()]
```

```python
# Longest Ride in distance
rides.loc[rides['distance'].idxmax()]
```

```python
sns.set(style="ticks", context="talk")
sns.regplot(x='distance', y = 'average_speed', data = rides).set_title("Average Speed vs Distance")
```

```python
sns.set(style="ticks", context="talk")
sns.regplot(x='distance', y = 'average_speed', data = runs).set_title("Average Speed vs Distance")
```
